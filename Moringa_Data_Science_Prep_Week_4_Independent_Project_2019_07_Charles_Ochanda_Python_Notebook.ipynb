# 1.6 Answering Research Questions

## 1.6.1 Bluecar counter 

```
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

# positive value mean a car was returned

# Negative value means a car was taken

# select city == Paris

car_paris = car_1[car_1['City'] =='Paris'] 

car_paris.head()



```

```
# Q1

# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

car_paris[car_paris['Bluecar_Diff'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```
# Q2

# What is the most popular hour for returning cars?

car_paris[car_paris['Bluecar_Diff'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```

# Q3

# What station is the most popular?

  # Overall?

car_paris[(car_paris['Kind'] == 'STATION') & (car_paris['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending= False)
```

```

# Q4

# What station is the most popular? At the most popular picking hour?
  
car_paris[(car_paris['Kind'] == 'STATION') & (car_paris['Status'] == 'ok') & (car_paris['hour'] == 4)].groupby('Public name')['Public name'].count().sort_values(ascending= False)


   
```

```
# Q5

#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?  Overall?

car_paris[(car_paris['Kind'] == 'STATION') & (car_paris['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)

# from the dataframe 75015 == Paris/Houdart de Lamotte/2 therefore NO it doesn't
  
    
```

```
# Q6

#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?

# At the most popular picking hour?

car_paris[(car_paris['Kind'] == 'STATION') & (car_paris['Status'] == 'ok') & (car_paris['hour'] == 4)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
 
```
